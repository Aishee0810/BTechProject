```python
def face_bounding_box(image,cascade,eye_cascade):
    # Convert the image to grayscale as the face detector expects a grayscale image.
#     image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    # Use detectMultiScale to detecting faces at multiple scales.
    faces = cascade.detectMultiScale(image, scaleFactor=1.2, minNeighbors=5);
#     print(faces)
    i = 0
    for (x, y, w, h) in faces:
        if w == 56 or w == 66:
            continue
        cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 3)
        roi_color = image[y:y+h, x:x+w]
        plt.subplot(121),plt.imshow(roi_color)
        plt.show()
#         cv2.imwrite('F:/academic_project/project/HAAR/AB'+str(i)+'.jpg',roi_color)
        i += 1
    plt.subplot(121),plt.imshow(image)
    plt.show()
    cv2.imwrite('F:/academic_project/project/HAAR/r2.jpg',image)
#         eyes = eye_cascade.detectMultiScale(roi_color, scaleFactor=1.3, minNeighbors=5)
#         for (ex,ey,ew,eh) in eyes:
#             cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(255,0,0),2)
#             plt.subplot(121),plt.imshow(roi_color)
#             plt.show()
#         cv2.imwrite('F:/academic_project/project/HAAR/'+name+'/'+img1,roi_color)
#     cv2.imwrite('F:/academic_project/project/HAAR/1s.jpg',image)
```
